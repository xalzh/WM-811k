```python
import pandas as pd
import numpy as np
import random
```

```python
try:
    df = pd.read_pickle("dataset\LSWMD.pkl")
except FileNotFoundError:
    print("Please download LSWMD.pkl dataset for WM-811k and place to the dataset folder")
print(df)
```

```python
df.sample(20)
```

```python
# Initialize dictionaries
train_data = {'Center': [], 'Donut': [], 'Edge-Loc': [], 'Edge-Ring': [], 'Loc': [], 'Random': [], 'Scratch': [], 'Near-full': [], 'none': []}
test_data = {'Center': [], 'Donut': [], 'Edge-Loc': [], 'Edge-Ring': [], 'Loc': [], 'Random': [], 'Scratch': [], 'Near-full': [], 'none': []}

# Process the dataset
for index, row in df.iterrows():
    wafer_map = np.array(row['waferMap'])

    # Ignore rows with empty labels and failure types
    if row['trianTestLabel'].size == 0 or row['failureType'].size == 0:
        continue

    label = row['trianTestLabel'][0][0]
    failure_type = row['failureType'][0][0]

    if label == 'Training':
        train_data[failure_type].append(wafer_map)
    elif label == 'Test':
        test_data[failure_type].append(wafer_map)
    else:
        print(f"Unexpected label: {label}")
```

```python
import matplotlib.pyplot as plt

def plot_data_summary(data_dict, dataset_name):
    print(f"{dataset_name} data summary:")
    total_count = 0
    for failure_type, samples in data_dict.items():
        count = len(samples)
        total_count += count
        print(f"  {failure_type}: {count} samples")
    print(f"Total samples in {dataset_name}: {total_count}\n")

    labels = list(data_dict.keys())
    sizes = [len(samples) for samples in data_dict.values()]

    # Pie chart
    plt.figure(figsize=(10, 10))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title(f"{dataset_name} Data Distribution - Pie Chart")
    plt.show()

    # Bar plot
    plt.figure(figsize=(12, 6))
    plt.bar(labels, sizes)
    plt.ylabel("Number of Samples (log scale)")
    plt.yscale('log')
    plt.title(f"{dataset_name} Data Distribution - Bar Plot")
    plt.show()

plot_data_summary(train_data, "Training")
plot_data_summary(test_data, "Testing")
```

```python
# Helper function to print one wafer map image array and its shape from the dictionary
def print_one_image(data_dict, failure_type):
    if data_dict[failure_type]:
        wafer_map_image = np.array(data_dict[failure_type][0])
        print(f"First wafer map image array for failure type '{failure_type}':")
        print(wafer_map_image)
        print(f"\nShape of the wafer map image array: {wafer_map_image.shape}")
    else:
        print(f"No samples available for failure type '{failure_type}'")

print_one_image(train_data, 'Center')
```

Data Preprocessing
1. Undersampling the majority class ('none').
2. Oversampling minority classes using SMOTE (requires imblearn library).
3. Resizing images to a fixed size (in this case, the average dimensions).

```python
from skimage.transform import resize
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

def preprocess_data(data_dict, size):
    preprocessed_data = {}
    for failure_type, samples in data_dict.items():
        resized_images = [resize(image, size, mode='reflect', anti_aliasing=True) for image in samples]
        preprocessed_data[failure_type] = resized_images
    return preprocessed_data

def apply_smote_to_data(data_dict):
    smote = SMOTE()
    all_image_vectors = []
    all_labels = []
    for failure_type, samples in data_dict.items():
        if failure_type != 'none':
            image_vectors = [image.flatten() for image in samples]
            labels = [failure_type] * len(image_vectors)
            all_image_vectors.extend(image_vectors)
            all_labels.extend(labels)
    oversampled_vectors, oversampled_labels = smote.fit_resample(all_image_vectors, all_labels)
    oversampled_data = {key: [] for key in data_dict.keys()}
    for vector, label in zip(oversampled_vectors, oversampled_labels):
        oversampled_image = np.array(vector).reshape(*target_size)
        oversampled_data[label].append(oversampled_image)
    # Add the 'none' class back to the oversampled data
    oversampled_data['none'] = data_dict['none']
    return oversampled_data

def balance_test_data(data_dict, samples_per_category=1000):
    balanced_data = {}
    for failure_type, samples in data_dict.items():
        if len(samples) >= samples_per_category:
            balanced_samples = random.sample(samples, samples_per_category)
        else:
            balanced_samples = samples
        balanced_data[failure_type] = balanced_samples
    return balanced_data



# Set the target image size (average dimensions, in this case)
target_size = (32, 32)

preprocessed_train_data = preprocess_data(train_data, target_size)
preprocessed_test_data = preprocess_data(test_data, target_size)

oversampled_train_data = apply_smote_to_data(preprocessed_train_data)
balanced_test_data = balance_test_data(preprocessed_test_data, samples_per_category=1000)

# Print the summary of the preprocessed data
plot_data_summary(oversampled_train_data, "Oversampled Training")
plot_data_summary(balanced_test_data, "Preprocessed Testing")
```
